```
def beziercurve(t,p1,p2,p3,p4):
  P = (p1*((1-t)**3))+(3*t*(((1-t)**2)*p2))+((3*(1-t))*((t**2)*p3))+((t**3)*p4)
  return P
```

```
def length(t):
  L = []
  difflen = []
  for i in range(1,202):
     L1x = beziercurve(0.005*(i-1),bezier[1]['A']['x'],bezier[1]['B']['x'],bezier[1]['C']['x'],bezier[1]['D']['x']) 
     L1y = beziercurve(0.005*(i-1),bezier[1]['A']['y'],bezier[1]['B']['y'],bezier[1]['C']['y'],bezier[1]['D']['y'])
     L2x = beziercurve(0.005*(i),bezier[1]['A']['x'],bezier[1]['B']['x'],bezier[1]['C']['x'],bezier[1]['D']['x']) 
     L2y = beziercurve(0.005*(i),bezier[1]['A']['y'],bezier[1]['B']['y'],bezier[1]['C']['y'],bezier[1]['D']['y'])
     if (L1x<0) :
       L1x = -L1x
     if (L1y<0) :
       L1y= -L1y 
     if (L2x<0) :
       L2x = -L2x
     if (L2y<0) :
       L2y= -L2y
     if ((L1x+L1y)-(L2x+L2y)<0):
       c = (L1x+L1y)-(L2x+L2y)
       c = -c
     else:
       c = (L1x+L1y)-(L2x+L2y)
     difflen.append(c)
  total = sum(difflen) # arc length
  L.append(total)
  L.append(total*(t-0))  #s(t)
  return L,difflen
```

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

N = 40
t = np.linspace(0,1,N)
x = 2*np.cos(6*np.pi*t)
y = 2*np.sin(6*np.pi*t)

dt = np.ones(N)
dx = -12*np.pi*np.sin(6*np.pi*t)
dy = 12*np.pi*np.cos(6*np.pi*t)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(t, x, y, 'gray')
ax.quiver3D(t, x, y, dt, dx, dy,length=1, normalize=True)
ax.scatter3D(t, x, y, c=y, cmap='hsv')
ax.set_xlabel("t axis")
ax.set_ylabel("x axis")
ax.set_zlabel("y axis")
plt.show()

normalizer = np.sqrt(dx*dx+dy*dy+dt*dt)
t = t/normalizer
x = x/normalizer
y = y/normalizer

dt = dt/normalizer
dx = dx/normalizer
dy = dy/normalizer

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(t, x, y, 'gray')
ax.quiver3D(t, x, y, dt, dx, dy)
ax.scatter3D(t, x, y, c=y, cmap='hsv')
ax.set_xlabel("t axis")
ax.set_ylabel("x axis")
ax.set_zlabel("y axis")
plt.show()
```
